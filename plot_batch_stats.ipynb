```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV
df = pd.read_csv("batch_stats.csv")

# Plot loss vs vertex count
plt.figure(figsize=(5, 5))
plt.scatter(df["vertex_count"], df["loss"], alpha=0.5)
plt.xlabel("Vertex Count")
plt.ylabel("Loss")
plt.title("Loss vs Vertex Count")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot loss vs node count
plt.figure(figsize=(5, 5))
plt.scatter(df["node_count"], df["loss"], alpha=0.5, color='orange')
plt.xlabel("Node Count")
plt.ylabel("Loss")
plt.title("Loss vs Node Count")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot loss vs node count
plt.figure(figsize=(5, 5))
plt.scatter(df["epoch"], df["loss"], alpha=0.5, color='green')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss vs Node Count")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot loss vs node count
plt.figure(figsize=(5, 5))
plt.scatter(df["batch_idx"], df["loss"], alpha=0.5, color='red')
plt.xlabel("Batch IDX")
plt.ylabel("Loss")
plt.title("Loss vs Node Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```
